```python
from nn import ActivationLayer
from main import show_result

def get_result(X, true_y, pred_y):
    hit = 0
    final_y = []
    for a, b in zip(true_y, pred_y):
        predict_label = (1 if b > 0.5 else 0)
        final_y.append(predict_label)
        if a[0] == predict_label:
            hit += 1

    show_result(X, true_y, final_y)
    print(hit/len(X))

def add_layer(network, layer, activation='sigmoid'):
    network.add(layer)
    activation_layer = ActivationLayer()
    activation_layer.set_activation_func(activation)
    network.add(activation_layer)
```

```python
from nn import NN, FCLayer
from main import generate_data

xor_network = NN()
add_layer(xor_network, FCLayer(2, 4), activation='tahh')
add_layer(xor_network, FCLayer(4, 4), activation='tahh')
add_layer(xor_network, FCLayer(4, 1), activation='sigmoid')
xor_network.set_loss_func('mse')

X, Y = generate_data('XOR')
xor_network.train(X, Y, epochs=5000, learning_rate=0.001, batch_size=21, verbose=True)
predict_Y = xor_network.predict(X)
get_result(X, Y, predict_Y)
print(predict_Y)
print(Y)
```

```python
from nn import NN, FCLayer
from main import generate_data


linear_network = NN()
linear_network.set_loss_func('mse')
add_layer(linear_network, FCLayer(2, 3), activation='sigmoid')
add_layer(linear_network, FCLayer(3, 1), activation='sigmoid')


X, Y = generate_data('linear')
linear_network.train(X, Y, epochs=30, learning_rate=0.1, batch_size=1, verbose=True)
predict_Y = linear_network.predict(X, batch_size=1)
get_result(X, Y, predict_Y)

```
